```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
ksize = 5  #50, #10
sigma = 3 #Large sigma on small features will fully miss the features.
theta = 1*np.pi/4  #/4 shows horizontal 3/4 shows other horizontal.
lamda = 1*np.pi /4  #1/4 works best for angled.
gamma=0.4  #Value of 1 defines spherical. values close to 0 has high aspect ratio
#Value of 1, spherical may not be ideal as it picks up features from other regions.
phi = 0
```

```python
kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)
plt.imshow(kernel)
```

```python
img = cv2.imread('/content/drive/MyDrive/zebra.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
fimg = cv2.filter2D(img, cv2.CV_8UC3, kernel)

kernel_resized = cv2.resize(kernel, (400, 400))                    # Resize image
cv2_imshow( kernel_resized)
cv2_imshow( img)
cv2_imshow( fimg)
```
